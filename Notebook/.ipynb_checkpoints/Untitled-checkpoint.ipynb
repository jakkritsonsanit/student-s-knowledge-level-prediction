```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load data

```python
train_raw_data = pd.read_excel(
            "../Data/Data_User_Modeling_Dataset_Hamdi Tolga KAHRAMAN.xls",
            sheet_name="Training_Data",
            usecols=range(0, 6))
```

```python
test_raw_data = pd.read_excel(
            "../Data/Data_User_Modeling_Dataset_Hamdi Tolga KAHRAMAN.xls",
            sheet_name="Test_Data",
            usecols=range(0, 6))
```

#### Concat data for preparing


Flag the data before concaternate

```python
train_raw_data['flag'] = "train"
test_raw_data['flag'] = "test"
```

```python
dataset = pd.concat([train_raw_data, test_raw_data])
```

```python
dataset.tail(10)
```

__Data Dict__
 - STG (The degree of study time for goal object meterails)
 - SCG (The degreee of repetition number of user for hoal object meterails)
 - STR (The degree of study time of user for related objects with goal object)
 - LRP (The exam performance of user for relatred objects with goal object)
 - PEG (The exam performance of user for goal objects)
 - UNS (The knowledge level of user)

```python
dataset.hist(figsize=(10,10), xrot=45)
plt.show()
```

#### Categorical distribution

```python
for col in dataset.select_dtypes('object'):
    sns.countplot(y=col, data=dataset)
    plt.show()
```

#### Segment and group by the target feature

```python
for col in dataset.select_dtypes('float64'):
    sns.boxenplot(y = dataset[' UNS'].astype('category'),
                 x = col,
                 data=dataset)
    plt.show()
```

Found that "very_low" level has two variant of typing

#### Check the correlation

```python
corr = dataset.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, cmap="RdBu_r", annot=True, vmax=1, vmin=-1)
plt.show()
```
